```python
# for loading, manipulating and cleaning the imported dataset
import pandas as pd
# for fast mathematical operations and handling numerical missing values
import numpy as np
# matplotlib's plotting tool to create basic visualisations like charts, lines, and figures
import matplotlib.pyplot as plt
# (built on top of matplotlib) to create prettier, advanced statistical graphs like boxplots
import seaborn as sns
```

```python
# Loads the messy CSV file into a pandas DataFrame (a table structure called 'df') for analysis
# pd.read_csv(): The function from the pandas library used to read CSV files.
df = pd.read_csv("crime_incidents_messy.csv")
```

```python
# Generates summary statistics for ALL columns (numerical and text) to reveal averages, ranges, unique values, and frequent categories
df.describe(include = 'all')
```

```python
# Counts how many total rows in the DataFrame are exact duplicates of another row
# .sum(): Adds up all the True values (since Python treats True as 1), giving you the exact total count of duplicate rows.
duplicate_count = df.duplicated().sum()

print(f"number of duplicated rows: {duplicate_count}")
```

```python
# Removes exact duplicate rows (keeping the very first one) and resets the row index so there are no missing numbers
df = df.drop_duplicates(keep = 'first').reset_index(drop=True)
```

```python
# Clean column names: strip edge whitespace, convert to lowercase, and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
# Prints a summary of the DataFrame, including total rows, column names, non-null counts, and data types (dtypes)
df.info()
```

```python
# Strip trailing/leading spaces and convert all text in 'crime_type' to lowercase
df['crime_type'] = df['crime_type'].str.strip().str.lower()
# Strip trailing/leading spaces and convert all text in 'case_status' to lowercase
df['case_status'] = df['case_status'].str.strip().str.lower()
# Print an array of all distinct crime categories to check for typos and duplicates
print(df['crime_type'].unique())
```

```python
# 1. First, squash multiple internal spaces down to single spaces and strip edges
df['crime_type'] = df['crime_type'].str.replace(r'\s+', ' ', regex=True).str.strip()
# 2. Clean dictionary (no trailing spaces, correct targets)
typo_map = {
    'asslt': 'assault', 
    'homocide': 'homicide', 
    'murder': 'homicide',
    'domestc violence': 'domestic violence', 
    'dv': 'domestic violence', 
    'dom. violence': 'domestic violence',
    'b&e': 'breaking & entering', 
    'narcotics': 'drug offense', 
    'drug offence': 'drug offense', 
    'drugs': 'drug offense',
    'burglry': 'burglary', 
    'fraudulent activity': 'fraud', 
    'scam': 'fraud', 
    'deception': 'fraud', 
    'online fraud': 'fraud',
    'arsen': 'arson', 
    'fire setting': 'arson',
    'trespass': 'trespassing', 
    'tresspassing': 'trespassing',
    'cyber crime': 'cybercrime', 
    'sexual assualt': 'sexual assault', 
    'sex assault': 'sexual assault',
    'sa': 'sexual assault',
    'drunk driving': 'dui', 
    'duii': 'dui', 
    'd.u.i': 'dui', 
    'dwi': 'dui', 
    'd.u.i.': 'dui',
    'robbry': 'robbery', 
    'roberry': 'robbery',
    'vandlism': 'vandalism', 
    'kidnaping': 'kidnapping', 
    'stealing': 'theft'
}
# 3. Apply replacement once
df['crime_type'] = df['crime_type'].replace(typo_map)
```

```python
# Prints the list of all unique crime categories remaining in the 'crime_type' column to verify that typos have been successfully cleaned
print(df['crime_type'].unique())
```

```python
# Prints a summary of the DataFrame, including total rows, column names, non-null counts, and data types (dtypes)
df.info()
```

```python
# Cleans the 'property_loss_usd' column by stripping out currency symbols ($), commas (,), and text, keeping ONLY numbers and decimal points
df['property_loss_usd'] = df['property_loss_usd'].astype(str).str.replace(r'[^\d]', '', regex=True)
```

```python
# Converts the 'property_loss_usd' column to numeric floats, turning any unparseable values into NaN (missing data)
# pd.to_numeric(...): Pandas function that transforms string representations of numbers (like '1250.50') into actual numerical numbers (1250.50) so you can perform calculations, sums, and averages.
# errors='coerce': The safety switch. If pandas runs into a value it can't convert into a number (like 'unknown', 'N/A', or empty strings ''), instead of crashing your code with a ValueError, it safely converts that bad value into NaN (Not a Number / missing value).
df['property_loss_usd'] = pd.to_numeric(df['property_loss_usd'], errors='coerce')
```

```python
# Print the data type of the 'property_loss_usd' column to verify it successfully converted to a float (numeric)
print("Column Type:", df['property_loss_usd'].dtype)
# Print a header for readability in the output
print("\nFirst 5 values after cleaning:")
# Display the first 5 rows of the cleaned numeric column to double-check the values
print(df['property_loss_usd'].head())
```

```python
# Convert the 'incident_datetime' column into standardized pandas Datetime objects, handling mixed structures and turning invalid dates into NaT
df['incident_datetime'] = pd.to_datetime(df['incident_datetime'], errors='coerce', format='mixed')

# Print the first 5 entries to confirm the dates now use the standard YYYY-MM-DD HH:MM:SS layout
print(df['incident_datetime'].head())
```

```python
# Fill missing values (NaN) in the 'victim_gender' column with the placeholder text 'unknown'
df['victim_gender'] = df['victim_gender'].fillna('unknown')

# Fill missing values (NaN) in the 'suspect_gender' column with the placeholder text 'unknown'
df['suspect_gender'] = df['suspect_gender'].fillna('unknown')

# Replace missing values (NaN) in 'victim_age' with the median age of all victims to prevent data distortion from extreme outliers
df['victim_age'] = df['victim_age'].fillna(df['victim_age'].median())

# Replace missing values (NaN) in 'suspect_age' with the median age of all suspects
df['suspect_age'] = df['suspect_age'].fillna(df['suspect_age'].median())

# Verify the changes by printing the count of missing (NaN) values remaining in these four specific columns
print(df[['victim_gender','suspect_gender', 'victim_age', 'suspect_age']].isna().sum())
```
